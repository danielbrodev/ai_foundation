```python
import numpy as np


arr = np.array([1,2,3,4,5])
print('numpy_array:',arr)

arr.shape
```

```python
arr2 = np.array([[1,2,3,4,5],[4,6,7,8,6]])
print('numpy_array2:',arr2)

arr2.shape
```

```python
# dtype()
a = np.array([1,2,3,4])
print(a.dtype)
```

```python
# reshape()
a = np.array([1,2,3,4,5,6])
b = a.reshape(2,3)
print('shaped_array:',b)
```

```python
#astpye()
a = np.array([1,2,3])
b = a.astype(float)
print('float:',b)
```

```python
import numpy as np

# np.zeros() np.ones() => dummy data
zer = np.zeros((2, 3))
on = np.ones((2, 3))

print('zeros:', zer)
print('ones:', on)
```

```python
# ndim
a = np.array([[1,2],[3,4]])

print('ndim:',a.ndim)
```

```python
# arrange()
a = np.arange(1,6)
print("arange:",a)
```

```python
# sum(), mean(), max(), min()
a = np.array([1,2,3,4])

print("sum:", a.sum())
print("mean:", a.mean())
print("max:", a.max())
print("min:", a.min())
```

```python
# flatten()

a = np.array([[1,2],[3,4]])
print("flatten:",a.flatten())
```

```python
# T (Transpose)

a = np.array([[1,2],[3,4]])
print("transpose:",a.T)
```

```python
# dot() @

a = np.array([[1,2],[3,4]])
b = np.array([[2,0],[1,2]])
print("dot/multiplication:",a.dot(b))
```

```python
import numpy as np

# random.rand()
a = np.random.rand(2,3) # => 0.0 and 1.0 / 0.0 included / 1.0 excluded
print("random rand:", a)

```
